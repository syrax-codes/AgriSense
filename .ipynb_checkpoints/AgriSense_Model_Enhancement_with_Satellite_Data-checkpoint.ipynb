# AgriSense Project Part 2: Satellite Data Analysis
This notebook analyzes the NDVI time-series data from Karnataka districts to perform feature engineering, anomaly detection, and forecasting using Satellite Data

## 1. Setup: Importing Libraries,Data Preprocessing and Initial Visualization(Initial Analysis)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

# Configure settings
warnings.filterwarnings('ignore')
sns.set_style('whitegrid')

# Load the new dataset
try:
    df_sat = pd.read_csv('satellite_data_all_districts.csv')
    print("Dataset 'satellite_data_all_districts.csv' loaded successfully.")
except FileNotFoundError:
    print("Error: The file 'satellite_data_all_districts.csv' was not found.")
    exit()

# --- 1. Initial Data Inspection ---
print("\n## 1. Initial Data Inspection")
print("\nFirst 5 rows of the dataset:")
print(df_sat.head())

print("\nShape of the dataset (rows, columns):", df_sat.shape)

print("\nData types and non-null counts:")
df_sat.info()

# --- 2. Descriptive Statistics ---
print("\n## 2. Descriptive Statistics for Numerical Columns")
print(df_sat.describe())

# --- 3. Check for Missing Values & Duplicates ---
print("\n## 3. Data Quality Check")
print("\nMissing values per column:")
print(df_sat.isnull().sum())
print(f"\nNumber of duplicate rows: {df_sat.duplicated().sum()}")
```

# Data Cleaning and Visualization
- Time-Series Plot: You will see the "heartbeat" of the agricultural seasons. NDVI values will rise during growing seasons (like the monsoon) and fall during dry seasons, showing the cycle of vegetation growth.
- Box Plot: This gives you a clear comparison of the districts. Districts with higher and wider boxes (like the coastal and Malnad regions) are generally greener and have more variability, while those with lower and narrower boxes are typically drier.

```python
# 1. Data Cleaning and Preprocessing 
print("## 1. Cleaning Data...")
df_sat.drop_duplicates(inplace=True)
df_sat['date'] = pd.to_datetime(df_sat['date'])
print("Data cleaning complete.")

# 2. Data Visualization 
print("\n## 2. Generating Visualizations...")

# a) Time-Series Plot for Sample Districts
plt.figure(figsize=(15, 7))
sample_districts = df_sat['District'].unique()[:4] 
for district in sample_districts:
    district_data = df_sat[df_sat['District'] == district].sort_values('date')
    plt.plot(district_data['date'], district_data['ndvi'], label=district)

plt.title('NDVI Time Series for Sample Districts', fontsize=16)
plt.xlabel('Date')
plt.ylabel('NDVI (Vegetation Index)')
plt.legend()
plt.grid(True)
plt.savefig('ndvi_timeseries.png')
plt.show()

# b) Box Plot of NDVI by District
plt.figure(figsize=(18, 8))
sorted_districts = df_sat.groupby('District')['ndvi'].median().sort_values().index
sns.boxplot(x='District', y='ndvi', data=df_sat, order=sorted_districts, palette='summer')
plt.title('Distribution of NDVI Across Districts', fontsize=16)
plt.xlabel('District')
plt.ylabel('NDVI (Vegetation Index)')
plt.xticks(rotation=90)
plt.tight_layout()
plt.savefig('ndvi_boxplot_by_district.png')
plt.show()
```

# Feature Engineering and Merging
This code will produce a new file named master_yield_satellite_data.csv. This is your new, powerful master dataset that combines agricultural statistics with satellite-derived vegetation data.

```python
# 1. LOAD AND CLEAN SATELLITE DATA 
print("## 1. Loading and Cleaning Satellite Data...")
try:
    df_sat = pd.read_csv('satellite_data_all_districts.csv')
    df_sat.drop_duplicates(inplace=True)
    df_sat['date'] = pd.to_datetime(df_sat['date'])
    df_sat['Year'] = df_sat['date'].dt.year
    df_sat['Month'] = df_sat['date'].dt.month
    print("Satellite data loaded and prepped.")
except FileNotFoundError:
    print("Error: satellite_data_all_districts.csv not found.")
    exit()

# 2. LOAD AND CLEAN YIELD DATA
print("\n## 2. Loading and Cleaning Historical Yield Data...")
try:
    raw_df_yield = pd.read_csv('historical_data_2010-2020.csv')
    
    # Logic to map each column to its correct season
    headers = pd.read_csv('historical_data_2010-2020.csv', nrows=1)
    season_map = {}
    current_season = None
    for col in headers.columns:
        # The season is in the first row of the raw file
        value = headers[col].iloc[0]
        if isinstance(value, str) and value in ['Kharif', 'Rabi', 'Summer']:
            current_season = value
        season_map[col] = current_season

    # Now proceed with the rest of the cleaning
    raw_df_yield['State'].ffill(inplace=True)
    raw_df_yield['District'].ffill(inplace=True)
    raw_df_yield['Year'].ffill(inplace=True)
    raw_df_yield = raw_df_yield.drop([0, 1]).reset_index(drop=True)

    tidy_data = []
    for i in range(3, raw_df_yield.shape[1], 3):
        if i + 2 >= raw_df_yield.shape[1]: continue
        
        crop_name_col = raw_df_yield.columns[i]
        header_info = pd.read_csv('historical_data_2010-2020.csv', nrows=1).columns[i]
        crop_name = header_info.split('/')[0].strip()
        
        # Get the season using our map
        season = season_map[crop_name_col]

        area_col, prod_col, yield_col = raw_df_yield.columns[i], raw_df_yield.columns[i+1], raw_df_yield.columns[i+2]
        crop_df = raw_df_yield[['State', 'District', 'Year', area_col, prod_col, yield_col]].copy()
        crop_df['Crop'] = crop_name
        crop_df['Season'] = season # Add the season column
        
        crop_df.rename(columns={
            area_col: 'Area (Hectare)', prod_col: 'Production (Tonnes)', yield_col: 'Yield (Kg/Ha)'
        }, inplace=True)
        tidy_data.append(crop_df)

    df_yield = pd.concat(tidy_data, ignore_index=True)
    df_yield.dropna(subset=['Area (Hectare)', 'Production (Tonnes)', 'Yield (Kg/Ha)', 'Season'], how='all', inplace=True)
    for col in ['Area (Hectare)', 'Production (Tonnes)', 'Yield (Kg/Ha)']:
        df_yield[col] = pd.to_numeric(df_yield[col].astype(str).str.replace(',', ''), errors='coerce')
    df_yield['District'] = df_yield['District'].str.split('. ').str[1]
    df_yield['District'] = df_yield['District'].str.upper()
    print("Yield data loaded and cleaned successfully (with seasons).")
except FileNotFoundError:
    print("Error: historical_data_2010-2020.csv not found.")
    exit()

#  3. FEATURE ENGINEERING FROM SATELLITE DATA 
print("\n## 3. Engineering New Features from NDVI Data...")
def get_season(month):
    if 6 <= month <= 10: return 'Kharif'
    elif month >= 11 or month <= 3: return 'Rabi'
    else: return 'Summer' # Standardized to 'Summer' to match yield file

df_sat['Season'] = df_sat['Month'].apply(get_season)

ndvi_features = df_sat.groupby(['District', 'Year', 'Season'])['ndvi'].agg(['mean', 'max', 'std', 'count']).reset_index()
ndvi_features.rename(columns={
    'mean': 'ndvi_mean', 'max': 'ndvi_max', 'std': 'ndvi_std', 'count': 'ndvi_reading_count'
}, inplace=True)
print("New NDVI features created.")

# 4. MERGE DATASETS 
print("\n## 4. Merging Yield Data with new NDVI Features...")
df_yield['Year'] = df_yield['Year'].str.slice(0, 4).astype(int)
df_master = pd.merge(df_yield, ndvi_features, on=['District', 'Year', 'Season'], how='left')

# 5. FINAL OUTPUT 
print("\nMerge complete! Here is the head of the new master DataFrame:")
print(df_master.head())
print("\nInfo of the new master DataFrame:")
df_master.info()

df_master.to_csv('master_yield_satellite_data.csv', index=False)
print("\nSuccessfully saved the combined dataset to 'master_yield_satellite_data.csv'")
```

# Retrain the Model with Satellite Features
- Here is the code to train and evaluate the enhanced model. This script will:
- Load your new master dataset.
- Prepare the data, now including the ndvi_ features.
- Train the XGBoost model using the best parameters we found earlier.
- Evaluate its performance.

## What to Look For
Compare the new R-squared (R^2) and MAE scores with the results from the previous model. We expect to see a noticeable improvement in these metrics, which will prove that integrating satellite data makes our AgriSense risk model more accurate and powerful. We are integrating a completely new and dynamic data source: near-real-time satellite imagery (NDVI). This isn't repeating the project; it's enhancing it.

```python
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error

# 1. LOAD THE ENRICHED DATASET 
try:
    df_master = pd.read_csv('master_yield_satellite_data.csv')
    print("Master dataset with satellite features loaded successfully.")
except FileNotFoundError:
    print("Error: 'master_yield_satellite_data.csv' not found. Please run the previous script first.")
    exit()

# 2. DATA PREPARATION 
# Drop rows where yield is NaN
df_master.dropna(subset=['Yield (Kg/Ha)'], inplace=True)

# Handle potential missing values in NDVI features (from the merge)
# We'll fill them with 0, assuming no satellite data means no significant vegetation.
ndvi_cols = ['ndvi_mean', 'ndvi_max', 'ndvi_std', 'ndvi_reading_count']
for col in ndvi_cols:
    df_master[col].fillna(0, inplace=True)

# Define features (X) and target (y)
# We include the new ndvi_ features and drop the leakage columns
features = df_master.drop(columns=['Yield (Kg/Ha)', 'Production (Tonnes)', 'Area (Hectare)', 'State'])
target = df_master['Yield (Kg/Ha)']

# One-Hot Encode categorical features
features_encoded = pd.get_dummies(features, columns=['District', 'Year', 'Season', 'Crop'], drop_first=True)

# 3. TRAIN-TEST SPLIT 
X_train, X_test, y_train, y_test = train_test_split(features_encoded, target, test_size=0.2, random_state=42)
print(f"\nTraining with {X_train.shape[1]} features.")

# 4. TRAIN THE ENHANCED XGBOOST MODEL
print("Training the enhanced XGBoost model...")
# Using the best parameters we found from hyperparameter tuning
best_params = {'learning_rate': 0.1, 'max_depth': 7, 'n_estimators': 200}
enhanced_model = xgb.XGBRegressor(objective='reg:squarederror', random_state=42, **best_params)

enhanced_model.fit(X_train, y_train)
print("Model training complete.")

# 5. EVALUATE THE ENHANCED MODEL
print("\n## Evaluating ENHANCED Model Performance...")
y_pred_enhanced = enhanced_model.predict(X_test)
r2_enhanced = r2_score(y_test, y_pred_enhanced)
mae_enhanced = mean_absolute_error(y_test, y_pred_enhanced)
rmse_enhanced = np.sqrt(mean_squared_error(y_test, y_pred_enhanced))

print(f"ENHANCED R-squared (R): {r2_enhanced:.4f}")
print(f"ENHANCED Mean Absolute Error (MAE): {mae_enhanced:.2f} Kg/Ha")
print(f"ENHANCED Root Mean Squared Error (RMSE): {rmse_enhanced:.2f} Kg/Ha")
```

# Anomaly Detection
- The resulting plot will show you the normal seasonal cycle for the selected district (the orange dotted line). The red dots highlight specific dates where the vegetation health was abnormally low compared to this trend.
- A cluster of these red dots could automatically trigger a risk alert in the AgriSense platform, prompting a closer look at that region and potentially protecting a lender's investment.

```python
#  1. LOAD AND PREPARE THE SATELLITE DATA 
try:
    df_sat = pd.read_csv('satellite_data_all_districts.csv')
    df_sat.drop_duplicates(inplace=True)
    df_sat['date'] = pd.to_datetime(df_sat['date'])
    df_sat.sort_values(['District', 'date'], inplace=True)
    print("Satellite data loaded and prepared.")
except FileNotFoundError:
    print("Error: 'satellite_data_all_districts.csv' not found.")
    exit()

# 2. PERFORM ANOMALY DETECTION FOR A SINGLE DISTRICT 

# ================================================================
# V V V V V V V V V CHOOSE A DISTRICT TO ANALYZE V V V V V V V V V
# ================================================================
district_to_analyze = 'BAGALKOT'
# ================================================================

print(f"\nAnalyzing anomalies for: {district_to_analyze}")
district_df = df_sat[df_sat['District'] == district_to_analyze].copy()

# Calculate the rolling mean and standard deviation
window_size = '30D' # 30-day rolling window
district_df['ndvi_rolling_mean'] = district_df.set_index('date')['ndvi'].rolling(window=window_size).mean().values
district_df['ndvi_rolling_std'] = district_df.set_index('date')['ndvi'].rolling(window=window_size).std().values

# Define the anomaly threshold (e.g., 2 standard deviations below the mean)
district_df['anomaly_threshold'] = district_df['ndvi_rolling_mean'] - (2 * district_df['ndvi_rolling_std'])

# Identify the anomalies
district_df['is_anomaly'] = district_df['ndvi'] < district_df['anomaly_threshold']
anomalies = district_df[district_df['is_anomaly']]

print(f"Found {len(anomalies)} potential anomaly points.")

# 3. VISUALIZE THE RESULTS 
plt.figure(figsize=(18, 8))
# Plot the actual NDVI data
plt.plot(district_df['date'], district_df['ndvi'], label='Actual NDVI', color='cornflowerblue', zorder=1)
# Plot the rolling average trend
plt.plot(district_df['date'], district_df['ndvi_rolling_mean'], label='30-Day Rolling Average', color='orange', linestyle='--', zorder=2)
# Plot the anomaly threshold
plt.plot(district_df['date'], district_df['anomaly_threshold'], label='Anomaly Threshold', color='gray', linestyle=':', zorder=2)
# Highlight the anomalies
plt.scatter(anomalies['date'], anomalies['ndvi'], color='red', s=100, label='Anomaly Detected', zorder=3, edgecolor='black')

plt.title(f'NDVI Anomaly Detection for {district_to_analyze}', fontsize=16)
plt.xlabel('Date')
plt.ylabel('NDVI (Vegetation Index)')
plt.legend()
plt.tight_layout()
plt.savefig('ndvi_anomaly_detection.png')
plt.show()
```

# NDVI Forecasting
- Main Forecast Plot: This is your key result.
    - Black dots are the actual historical NDVI values.
    - The dark blue line is the model's forecast.
    - The light blue shaded area represents the uncertainty interval. The model predicts the true value will likely fall within this range.

- Components Plot: This shows what the model has learned about your data.
    - Trend: Shows the long-term direction of the NDVI in that district.
    - Yearly: This is the most important part for agricultural data. It will show a clear wave-like pattern, representing the predictable rise and fall of "greenness" throughout the seasons each year.

```python
# PART 1: INSTALL PROPHET LIBRARY 
# This command installs the prophet library.
!pip install -q prophet

# --- PART 2: IMPORTS & DATA PREPARATION ---
import pandas as pd
from prophet import Prophet
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings('ignore')

# Load and prepare the satellite data
try:
    df_sat = pd.read_csv('satellite_data_all_districts.csv')
    df_sat.drop_duplicates(inplace=True)
    df_sat['date'] = pd.to_datetime(df_sat['date'])
    print("Satellite data loaded and prepared.")
except FileNotFoundError:
    print("Error: 'satellite_data_all_districts.csv' not found.")
    exit()

# PART 3: PREPARE DATA FOR PROPHET 

# ================================================================
# V V V V V V V V V CHOOSE A DISTRICT TO FORECAST V V V V V V V V V
# ================================================================
district_to_forecast = 'BAGALKOT'
# ================================================================

print(f"\nPreparing data for: {district_to_forecast}")
district_df = df_sat[df_sat['District'] == district_to_forecast].copy()

# Prophet requires specific column names: 'ds' for the date and 'y' for the value
prophet_df = district_df[['date', 'ndvi']].rename(columns={'date': 'ds', 'ndvi': 'y'})

# PART 4: TRAIN THE FORECASTING MODEL 
print("Training the Prophet forecasting model...")

# Initialize and fit the model
model = Prophet(yearly_seasonality=True, daily_seasonality=False, weekly_seasonality=False)
model.fit(prophet_df)
print("Model training complete.")

# --- PART 5: MAKE FUTURE PREDICTIONS ---
# Create a dataframe for future dates (e.g., the next 365 days)
future = model.make_future_dataframe(periods=365)
# Generate the forecast
forecast = model.predict(future)

print("\n## Generating Forecast Visualizations...")

# PART 6: VISUALIZE THE FORECAST 

# a) Plot the main forecast
fig1 = model.plot(forecast)
plt.title(f"NDVI Forecast for {district_to_forecast}", fontsize=16)
plt.xlabel("Date")
plt.ylabel("Predicted NDVI")
plt.savefig('ndvi_forecast.png')
plt.show()

# b) Plot the forecast components (trend and seasonality)
fig2 = model.plot_components(forecast)
plt.savefig('ndvi_forecast_components.png')
plt.show()
```
